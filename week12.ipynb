#실습1 : 서울의 일교차

기온 공공 데이터를 이용하여 서울에서 일교차가 가장 큰 날, 최저기온이 가장 낮은 날, 최고 기온이 가장 높은 날을 찾아 출력하시오. (20210101~20211231)

- 출력 형식

>일교차가 가장 큰 날  
날짜  
최저기온이 가장 낮은 날  
날짜  
최고기온이 가장 높은 날  
날짜  

```python
# 실습1

import csv
f = open('seoul_2021.csv', 'r')
data = csv.reader(f)
header = next(data)

max_dif = -999
max_temp = -999
min_temp = 999

max_dif_date = ''
max_temp_date = ''
min_temp_date = ''

for row in data:
    row_date = row[0]
    row_min = float(row[3])
    row_max = float(row[4])
    row_dif = row_max - row_min
    
    if(row_dif > max_dif):
        max_dif = row_dif
        max_dif_date = row_date
    
    if(row_max > max_temp):
        max_temp = row_max
        max_temp_date = row_date
        
    if(row_min < min_temp):
        min_temp = row_min
        min_temp_date = row_date
        
f.close()

print("일교차가 가장 큰 날")
print(max_dif_date)
print("최저기온이 가장 낮은 날")
print(min_temp_date)
print("최고기온이 가장 높은 날")
print(max_temp_date)       
        
```

#실습2 : 지하철 이용객

"지하철 노선별 역별 이용현황" 데이터를 이용하여 노선별로 가장 이용객이 많은 3개역과 가장 적은 3개 역을 출력하시오.(노선마다 출력할 것!)  
=> 2021년 12월 데이터 사용, 이용객 = 승차인원 + 하차인원

- 출력형식  

>이용객이 가장 많은 역 순위  
1.역명(이용객 수)  
2.역명(이용객 수)  
3.역명(이용객 수)  
이용객이 가장 적은 역 순위  
1.역명(이용객 수)  
2.역명(이용객 수)  
3.역명(이용객 수)  

```python
# 실습2

import csv
f = open('subwayfee_202112.csv','r')
data = csv.reader(f)
header = next(data)

is_first_line = True
before_line_name = ''
max_people = [0] * 3
min_people = [99999999] * 3
max_station = [''] * 3
min_station = [''] * 3
    
def print_data():
    print(before_line_name)
    print("이용객이 가장 많은 역 순위")
    for i in range(0,3):
        print("{0:d}.{1:s}({2:d})".format(i+1,max_station[i],max_people[i]))
    print("이용객이 가장 적은 역 순위")
    for i in range(0,3):
        print("{0:d}.{1:s}({2:d})".format(i+1,min_station[i],min_people[i]))
    print()


for row in data:
    line_name = row[1]
    station_name = row[3]
    people = int(row[4].replace(",","")) + int(row[5].replace(",",""))
    
    if is_first_line :
        is_first_line = False
        before_line_name = line_name
        
    elif before_line_name != line_name:
        print_data()
        max_people = [0] * 3
        min_people = [99999999] * 3
        max_station = [''] * 3
        min_station = [''] * 3
        before_line_name = line_name
                
    for i in range(0,3):
        if people > max_people[i]:
            max_people.insert(i, people)
            max_people.pop()
            max_station.insert(i, station_name)
            max_station.pop()
            break
    
    for i in range(0,3):
        if people < min_people[i]:
            min_people.insert(i, people)
            min_people.pop()
            min_station.insert(i, station_name)
            min_station.pop()
            break

print_data()
f.close()
```

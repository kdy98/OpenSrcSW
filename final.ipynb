```python
import csv

f = open('country_vaccinations.csv')
data = csv.reader(f)
header = next(data)

dic = dict()

max_value = 0
max_date = ""
min_value = 9999999999
min_date = ""

for row in data:
    if row[7] != '':
        if row[2] in dic:
            dic[row[2]] += float(row[7])
        else:
            dic[row[2]] = float(row[7])

            
for i in dic:
    if dic[i]>max_value:
        max_value = dic[i]
        max_date = i
        
    if dic[i]<min_value:
        min_value = dic[i]
        min_date = i
        
print("백신이 가장 많이 접종된 날")
print(max_date)
print("백신이 가장 적게 접종된 날")
print(min_date)
```

```python
import csv
import pandas as pd
import matplotlib.pyplot as plt

f = open('country_vaccinations.csv')
data = csv.reader(f)
header = next(data)

dic = dict()
vaccine = []

for row in data:
    vaccines = row[-3].split(",")
    country = row[0]
    for i in vaccines:
        if i in dic:
            if not country in dic[i]:
                dic[i].append(country)
        else:
            dic[i] = []

vcount = dict()

for i in dic:
    vcount[i] = len(dic[i])
    
#print(vcount)

xdata = vcount.keys()
ydata = []

for i in xdata:
    ydata.append(vcount[i])
    
a = pd.DataFrame(ydata,index=xdata,columns=["접종 국가 수"])
print(a)

plt.pie(ydata,labels=xdata)
plt.legend()
plt.show()
```
